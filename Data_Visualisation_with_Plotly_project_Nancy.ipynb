```python
subscribers_df = pd.read_csv('https://bit.ly/MobileDataset')
subscribers_df.head()
```

```python
subscribers_df.shape
```

```python
subscribers_df.dtypes
```

```python
# check for missing values in each column 

subscribers_df.isnull().sum()
```

```python
#  convert all values in our colums to lower case
#df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
# mobile_df.columns = mobile_df.columns.str.lower()
subscribers_df.columns = subscribers_df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
subscribers_df.tail()
```

```python
subscribers_df.columns = ['series_name', 'series_code', 'country_name', 'country_code', '1995', '1996', '1997', '1998','1999' ,'2000',' 2001',' 2002',' 2003',' 2004',' 2005',' 2006',' 2007 ','2008 ','2009 ','2010 ','2011 ','2012 ','2013',' 2014','2015', '2016']
subscribers_df.columns = subscribers_df.columns.str.strip()
```

```python
subscribers_df[subscribers_df.duplicated()]
sum(subscribers_df.duplicated())
```

No duplicates

```python
#convert coulumns to int 
yearly_columns = subscribers_df.columns[4:]
subscribers_df[yearly_columns] = subscribers_df[yearly_columns].apply(lambda x: x.replace('..', '0')).astype(float).astype('int')

yearly_columns
```

Dropping colums with zero values

```python
subscribers_df.drop(columns=['2016'], inplace=True)
subscribers_df.sample(10)
```

```python
subscribers_df['series_name'].unique()
subscribers_df.head()
```

```python
stats_columns = list(subscribers_df.columns[1:])
check_df = subscribers_df.groupby('series_name', as_index=False).sum()
check_df.head()
```

```python
check_df = check_df.transpose().reset_index()
check_df
```

```python
column_new  = check_df.iloc[0]
column_new
```

```python
check_df.columns = column_new
check_df.drop(index=0, inplace=True)
check_df
```

```python
new_column = check_df.columns[1:]
check_df = pd.melt(check_df, id_vars=['series_name'], value_vars='value_columns')
check_df.head(5)
```

```python
fig = px.line(
    check_df, 
    x = 'series_name',
    y = 'value',
    color = 0,

    template = "presentation"
  )

# Tweak and Label
fig.update_layout(
    title = {
        'text': ' KPIs trends years', 
        'x': 0.5,
        'font_size' : 15,
        'xanchor': 'center',
        'yanchor': 'top'},
    xaxis_title_text = 'Years',
    yaxis_title_text = 'number',
  ) 

# Display
fig.show()
```

```python
fig2 = px.line(
    check_df, 
    x = 'series_name',
    y = 'value',
    color = 0,

    template = "presentation"
  )

# Tweak and Label
fig2.update_layout(
    title = {
        'text': ' KPIs trends years', 
        'x': 0.5,
        'font_size' : 15,
        'xanchor': 'center',
        'yanchor': 'top'},
    xaxis_title_text = 'Years',
    yaxis_title_text = 'number',
  ) 

# Display
fig2.show()
```

```python
yearly_columns = list(subscribers_df.columns[2:])
country_trend_df = pd.melt(subscribers_df, id_vars=['series_name', 'country_name'], value_vars=yearly_columns)
cellular_df = country_trend_df[country_trend_df['series_name']=='Fixed broadband subscriptions']

fig3 = px.line(
    cellular_df, 
    x = 'variable',
    y = 'value',
    color = 'country_name',

    template = "simple_white"
  )

# Tweak and Label
fig3.update_layout(
    title = {
        'text': 'fixed broadband  growth  per country', 
        'x': 0.5,
        'font_size' : 15,
        'xanchor': 'center',
        'yanchor': 'top'},
    xaxis_title_text = 'Years',
    yaxis_title_text = 'fixed broadband ',
  ) 

# Display
fig3.show()
```

```python
mobile_cell_df = country_trend_df[country_trend_df['series_name']=='Mobile cellular subscriptions']

fig4 = px.line(
    mobile_cell_df, 
    x = 'variable',
    y = 'value',
    color = 'country_name',

    template = "plotly_dark"
  )

# Tweak and Label
fig4.update_layout(
    title = {
        'text': ' cellular subscriptions  per country', 
        'x': 0.5,
        'font_size' : 15,
        'xanchor': 'center',
        'yanchor': 'top'},
    xaxis_title_text = 'Years',
    yaxis_title_text = 'cellular subscriptions',
  ) 

# Display
fig4.show()
```
